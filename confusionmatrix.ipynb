```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Hard-coded confusion matrices
confusion_matrices = [
    np.array([[799, 0], [1, 27]]),
    np.array([[788, 5], [0, 34]]),
    np.array([[797, 0], [0, 30]]),
    np.array([[809, 2], [0, 16]]),
    np.array([[814, 0], [8, 5]]),
    np.array([[790, 0], [5, 32]])
]

# Diagnoses
diagnoses = ["1dAVb", "RBBB", "LBBB", "SB", "AF", "ST"]

# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Plot confusion matrices
for i, ax in enumerate(axes.flatten()):
    sns.heatmap(confusion_matrices[i], annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=["Not Present", "Present"],
                yticklabels=["Not Present", "Present"], ax=ax)
    ax.set_title(diagnoses[i])
    ax.set_xlabel('Predicted Label')
    ax.set_ylabel('Actual Label')

plt.tight_layout()
plt.show()
```

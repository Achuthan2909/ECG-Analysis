```python
import pandas as pd

# Assuming your data is stored in a DataFrame called 'df'
file_path = "/home/achu/Desktop/GitHubECGWork/Implementation_of_paper/ECG-Analysis/ECG-Analysis/data/annotations/gold_standard.csv"
df = pd.read_csv(file_path)
# Select the columns of interest for conditions
conditions_columns = ['1dAVb','RBBB','LBBB','SB','AF','ST']
# Create a new column 'num_diseases' representing the total number of diseases for each patient
df['num_diseases'] = df[conditions_columns].sum(axis=1)
# Initialize a dictionary to store the counts of each combination
combination_counts = {}
# Iterate over unique combinations of diseases
for combination in df[conditions_columns].apply(tuple, axis=1).unique():
    count = (df[conditions_columns].apply(tuple, axis=1) == combination).sum()
    combination_counts[combination] = count
# Iterate over unique combinations of diseases
for combination, count in combination_counts.items():
    # Filter the diseases that are True in the combination
    true_diseases = [condition for condition, is_true in zip(conditions_columns, combination) if is_true]
    # Display the counts for each combination
    print(f"Combination {', '.join(true_diseases)}: {count} patients")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming combination_counts is a dictionary with combination counts

# Function to categorize combinations based on the number of "True" values
def categorize_combinations(combination):
    num_true_values = sum(combination)
    if num_true_values == 0:
        return 'No True'
    elif num_true_values == 1:
        return '1 True'
    else:
        return 'Others'

# Categorize combinations and count patients
categorized_counts = {'Not Classified': 0, '1 True': 0, 'Other Combination': 0}
for combination, count in combination_counts.items():
    category = categorize_combinations(combination)
    categorized_counts[category] += count

# Pie chart for categorizations
labels = list(categorized_counts.keys())
sizes = list(categorized_counts.values())

# Pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("viridis", n_colors=len(labels)))
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
ax.set_title('Combinations with 1 True and No True')

plt.show()
```
